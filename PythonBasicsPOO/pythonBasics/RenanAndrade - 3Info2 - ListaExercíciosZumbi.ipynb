1) Faça um programa que peça dois números inteiros e imprima a soma desses dois números

```python
primeiroNumero = int(input("Digite o segundo número da soma: \n ─>"))
segundoNumero = int(input("Digite o primeiro número da soma: \n ─>"))

soma = primeiroNumero + segundoNumero
print("O total da soma é",soma)
```

2) Escreva um programa que leia um valor em metros e o exiba convertido em milímetros

```python
valorMetros = float(input("Digite p valor, em metros, a ser convertido em milímetros: \n ─>"))
fatorConversao = 1000
totalMilimetros = valorMetros * fatorConversao
print(valorMetros, 'metro(s) é equivalente a', totalMilimetros, 'milímetros')
```

3) Escreva um programa que leia a quantidade de dias, horas, minutos e segundos do usuário. Calcule o total em segundos.

```python
quantidadeSegundos = float(input("Quantidade de segundos do usuário: \n ─>"))

quantidadeMinutos = float(input("Quantidade de minutos do usuário: \n ─>"))
minutosParaSegundos = quantidadeMinutos * 60

quantidadeHoras = float(input("Quantidade de horas do usuário: \n ─>"))
horasParaSegundos = quantidadeHoras * 60 * 60

quantidadeDias = float(input('Quantidade de dias do usuário: \n ─>'))
diasParaSegundos = quantidadeDias * 24 * 60 * 60

totalSegundos = diasParaSegundos + minutosParaSegundos + horasParaSegundos + quantidadeSegundos
print("O usuário tem", totalSegundos, "segundos")
```

9) Escreva um programa que pergunte a quantidade de km percorridos por um carro alugado pelo usuário, assim como a quantidade de dias pelos quais o carro foi alugado. Calcule o preço a pagar, sabendo que o carro custa 60,00 reais por dia e 0,15 reais por km rodado.

```python
taxaPorKm = 0.15
taxaAluguel = 60.00

kmPercorridos = float(input('Quantidade de kilômetros percorridos com o carro: \n ─>'))
diasCarroAlugado = float(input('Quantidade de dias com o carro alugado: \n ─>'))
 
totalPreço = kmPercorridos * taxaPorKm + taxaAluguel * diasCarroAlugado
print("O total a ser pago é de", totalPreço, "reais")
```

10) Escreva um programa para calcular a redução do tempo de vida de um fumante. Pergunte a quantidade de cigarros fumados por dia e quantos anos ele já fumou. Considere que um fumante perde 10 minutos de vida a cada cigarro, calcule quantos dias de vida um fumante perderá. Exiba o total de dias.

```python
vidaPerdidaPorCigarro = 10 / 60 / 24 #já convertida para dias
cigarrosConsumidosPorDia = int(input("Quantos cigarros o indivíduo fuma por dia? \n ─> "))

anosParaDias = 365
anosFumando = float(input("Quantos anos esteve fumando? \n ─> "))

totalDiasPerdidos = vidaPerdidaPorCigarro * cigarrosConsumidosPorDia * anosParaDias * anosFumando
print("Esta pessoa perdeu", round(totalDiasPerdidos, 2), "dias da vida dela")
```
